# PARLIAMENTARY SEATS IN SINGLE OR LOWER CHAMBERS

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as sta

df = pd.read_csv('data.csv')

for i in range (2011, 2021, 1):
    
    x = df["COUNTRIES"]
    y = df[f"{i-1}"]
    
    s_deviation = y.std()
    
    maximum_val = y.max()
    minimum_val = y.min()
    Range_val = maximum_val - minimum_val
    
    variance_val = sta.variance(df[f"{i-1}"])

    plt.figure(figsize =(35, 20))
    plt.barh(x, y, height = 0.7)
    plt.title(f"NUMBER OF PARLIAMENTARY SEATS IN {i-1}", fontsize = 30)
    plt.xlabel('Number of Seats', fontsize = 30)
    plt.xticks(fontsize = 25)
    plt.yticks(fontsize = 20)
    plt.show()
    
    print(f'The Standard Deviation is {s_deviation}')
    print(f'The Range is {Range_val}')
    print(f'The Variance is {variance_val}')
```
